```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix

import pickle
```

```python
data = pd.read_csv('/kaggle/input/heart-disease-uci/heart.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
sns.countplot(x='sex', hue='target', data=data)
```

```python
labels = data['target']
features = data.drop('target', axis=1)
```

```python
labels.head()
```

```python
features.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.3)
```

## Logisitic Regression

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg_pred = log_reg.predict(X_test)
```

```python
np.array(y_test)
```

```python
log_reg_pred
```

```python
print(classification_report(y_test, log_reg_pred))
```

```python
print(confusion_matrix(y_test, log_reg_pred))
```

```python
with open('logistic_model.pkl', 'wb') as f:
    pickle.dump(log_reg, f)
```

## Random Forest

```python
SC = StandardScaler()
SC.fit(features)
sc_features = SC.transform(features)
```

```python
sc_features_df = pd.DataFrame(sc_features)
sc_features_df.columns = features.columns
```

```python
sc_features_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(sc_features, labels, test_size = 0.3)
```

```python
RFC = RandomForestClassifier()
RFC.fit(X_train, y_train)
rfc_pred = RFC.predict(X_test)
```

```python
np.array(y_test)
```

```python
rfc_pred
```

```python
print(classification_report(y_test, rfc_pred))
```

```python
print(confusion_matrix(y_test, rfc_pred))
```

```python
with open('RF_model.pkl', 'wb') as f:
    pickle.dump(RFC, f)
```

## SVC

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.3)
```

```python
param = {
    'C':[1,10,100,10000,100000, 1000000],
    'gamma': [0.0000001, 0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1]
}
```

```python
svc = GridSearchCV(SVC(), param, verbose=2)
```

```python
svc.fit(X_train, y_train)
```

```python
svc.best_estimator_
```

```python
svc_pred = svc.predict(X_test)
```

```python
np.array(y_test)
```

```python
svc_pred
```

```python
print(classification_report(y_test, svc_pred))
```

```python
print(confusion_matrix(y_test, svc_pred))
```

```python
with open('svm_clf_model.pkl', 'wb') as f:
    pickle.dump(svc, f)
```
